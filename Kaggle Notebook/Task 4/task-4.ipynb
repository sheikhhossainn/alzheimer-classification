```python
import os
import random
import time
import gc
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.layers import (
    Input, Conv2D, MaxPooling2D, BatchNormalization, Activation,
    Dense, Dropout, GlobalAveragePooling2D, GlobalMaxPooling2D,
    Reshape, Multiply, Add, Concatenate, Lambda
)
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import f1_score, precision_score, recall_score, confusion_matrix

# ==========================================
# 1. SCIENTIFIC LOCKS  (identical to Task 3)
# ==========================================
SEED = 42
os.environ['PYTHONHASHSEED'] = str(SEED)
os.environ['TF_DETERMINISTIC_OPS'] = '1'
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ==========================================
# 2. DUAL-GPU HARDWARE SETUP
# ==========================================
if 'strategy' not in locals():
    try:
        strategy = tf.distribute.MirroredStrategy()
        print(f"вњ… Dual-GPU Strategy Initialized. Replicas: {strategy.num_replicas_in_sync}")
    except Exception as e:
        print(f"вќЊ GPU initialization failed: {e}")
        strategy = tf.distribute.get_strategy()
else:
    print(f"вљЎ Strategy already active. Replicas: {strategy.num_replicas_in_sync}")

BATCH_SIZE = 32 * strategy.num_replicas_in_sync
AUTOTUNE   = tf.data.AUTOTUNE

# ==========================================
# 3. DATA LOADING & SORTING  (identical to Task 3)
# ==========================================
DATA_DIR = '/kaggle/input/datasets/uraninjo/augmented-alzheimer-mri-dataset-v2'
print("\nFetching Local File Paths...")
filepaths = []

for root, dirs, files in os.walk(DATA_DIR):
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            filepaths.append(os.path.join(root, file))

filepaths.sort()   # CRUCIAL FOR T-TEST MATCHING

labels  = [os.path.basename(os.path.dirname(path)) for path in filepaths]
df      = pd.DataFrame({'filepath': filepaths, 'label': labels})
class_names   = sorted(df['label'].unique())
class_indices = {name: idx for idx, name in enumerate(class_names)}
df['label_idx'] = df['label'].map(class_indices)

print(f"вњ… Successfully pooled and sorted {len(df)} images.")
print(f"Classes found: {class_indices}")

# ==========================================
# 4. DATA PIPELINE  (identical to Task 3)
# ==========================================
def decode_image(filepath, label):
    image_bytes = tf.io.read_file(filepath)
    image = tf.io.decode_image(image_bytes, channels=3, expand_animations=False)
    image = tf.image.resize(image, [224, 224])
    image = image / 255.0
    label = tf.one_hot(label, depth=len(class_names))
    return image, label

def decode_image_predict_only(filepath):
    image_bytes = tf.io.read_file(filepath)
    image = tf.io.decode_image(image_bytes, channels=3, expand_animations=False)
    image = tf.image.resize(image, [224, 224])
    image = image / 255.0
    return image

def build_dataset(dataframe, is_training=False):
    dataset = tf.data.Dataset.from_tensor_slices(
        (dataframe['filepath'].values, dataframe['label_idx'].values)
    )
    dataset = dataset.map(decode_image, num_parallel_calls=AUTOTUNE)
    if is_training:
        dataset = dataset.shuffle(buffer_size=2048, seed=SEED)
    dataset = dataset.batch(BATCH_SIZE).prefetch(AUTOTUNE)
    return dataset

def build_predict_dataset(dataframe):
    dataset = tf.data.Dataset.from_tensor_slices(dataframe['filepath'].values)
    dataset = dataset.map(
        decode_image_predict_only, num_parallel_calls=AUTOTUNE, deterministic=True
    )
    dataset = dataset.batch(BATCH_SIZE).prefetch(AUTOTUNE)
    return dataset


# ==========================================
# 5. CBAM ATTENTION MODULE
# ==========================================
# CBAM = Convolutional Block Attention Module
# Paper: "CBAM: Convolutional Block Attention Module" (Woo et al., ECCV 2018)
#
# How it works:
#   Step 1 вЂ” Channel Attention: asks "which feature maps matter?"
#             Squeezes spatial dims в†’ MLP в†’ sigmoid gate в†’ rescale channels
#   Step 2 вЂ” Spatial Attention: asks "which regions matter?"
#             Pool across channels в†’ conv в†’ sigmoid gate в†’ rescale spatial positions
#
# Net effect: the model learns to suppress irrelevant channels AND irrelevant
# spatial regions, focusing on discriminative brain regions for Alzheimer's.

def channel_attention(x, ratio=8):
    """
    Channel Attention: recalibrates 'which feature maps' are important.
    ratio controls the bottleneck size in the shared MLP.
    """
    channels = x.shape[-1]

    # Shared MLP weights вЂ” applied to both avg and max pooled features
    shared_dense_1 = Dense(channels // ratio, activation='relu',  use_bias=False)
    shared_dense_2 = Dense(channels,          activation='linear', use_bias=False)

    # Average pool path
    avg_pool = GlobalAveragePooling2D()(x)           # (B, C)
    avg_pool = Reshape((1, 1, channels))(avg_pool)   # (B, 1, 1, C)
    avg_out  = shared_dense_1(avg_pool)
    avg_out  = shared_dense_2(avg_out)

    # Max pool path
    max_pool = GlobalMaxPooling2D()(x)
    max_pool = Reshape((1, 1, channels))(max_pool)
    max_out  = shared_dense_1(max_pool)
    max_out  = shared_dense_2(max_out)

    # Combine + sigmoid gate
    attention = Add()([avg_out, max_out])
    attention = Activation('sigmoid')(attention)     # (B, 1, 1, C)

    return Multiply()([x, attention])                # rescale input channels


def spatial_attention(x, kernel_size=7):
    """
    Spatial Attention: recalibrates 'which spatial locations' are important.
    """
    # Pool across channel axis to get two 2D maps
    avg_pool = Lambda(lambda t: tf.reduce_mean(t, axis=-1, keepdims=True))(x)  # (B, H, W, 1)
    max_pool = Lambda(lambda t: tf.reduce_max(t,  axis=-1, keepdims=True))(x)  # (B, H, W, 1)

    concat    = Concatenate(axis=-1)([avg_pool, max_pool])                      # (B, H, W, 2)
    attention = Conv2D(1, kernel_size, padding='same', use_bias=False)(concat)
    attention = Activation('sigmoid')(attention)                                # (B, H, W, 1)

    return Multiply()([x, attention])                # rescale input spatially


def cbam_block(x, ratio=8, kernel_size=7):
    """
    Full CBAM: Channel Attention в†’ Spatial Attention (applied sequentially).
    Drop this after any Conv block вЂ” zero architecture changes needed elsewhere.
    """
    x = channel_attention(x, ratio=ratio)
    x = spatial_attention(x, kernel_size=kernel_size)
    return x


# ==========================================
# 6. ATTENTION CNN ARCHITECTURE
# ==========================================
# Identical to Task 3 except CBAM is inserted after every Conv block.
# This is the ONLY change вЂ” everything else (filters, LR, optimizer) is
# kept the same for a fair apples-to-apples comparison.

def build_attention_cnn_model():
    tf.keras.backend.clear_session()
    with strategy.scope():
        inputs = Input(shape=(224, 224, 3))

        # Block 1 + CBAM
        x = Conv2D(32, (3, 3), padding='same')(inputs)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)
        x = MaxPooling2D(pool_size=(2, 2))(x)
        x = cbam_block(x)                           # в†ђ CBAM after Block 1

        # Block 2 + CBAM
        x = Conv2D(64, (3, 3), padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)
        x = MaxPooling2D(pool_size=(2, 2))(x)
        x = cbam_block(x)                           # в†ђ CBAM after Block 2

        # Block 3 + CBAM
        x = Conv2D(128, (3, 3), padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)
        x = MaxPooling2D(pool_size=(2, 2))(x)
        x = cbam_block(x)                           # в†ђ CBAM after Block 3

        # Block 4 + CBAM
        x = Conv2D(256, (3, 3), padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)
        x = MaxPooling2D(pool_size=(2, 2))(x)
        x = cbam_block(x)                           # в†ђ CBAM after Block 4

        # Top Layers  (identical to Task 3)
        x = GlobalAveragePooling2D()(x)
        x = Dense(256, activation='relu')(x)
        x = Dropout(0.5)(x)
        outputs = Dense(len(class_names), activation='softmax')(x)

        model = Model(inputs=inputs, outputs=outputs)
        model.compile(
            optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
    return model


# ==========================================
# 7. TRAINING & EVALUATION LOOP
# ==========================================
K_FOLDS = 5
EPOCHS  = 50

skf = StratifiedKFold(n_splits=K_FOLDS, shuffle=True, random_state=SEED)

fold_accuracies    = []
fold_f1_scores     = []
fold_precisions    = []
fold_recalls       = []
fold_conf_matrices = []

print("\nрџљЂ STARTING 5-FOLD CROSS-VALIDATION (CBAM Attention CNN on T4 x2 GPUs)...")
start_time = time.time()

for fold, (train_val_idx, test_idx) in enumerate(skf.split(df['filepath'], df['label_idx'])):
    fold_num = fold + 1
    print(f"\n{'='*50}\nрџ§Є STARTING FOLD {fold_num} / {K_FOLDS}\n{'='*50}")

    test_df      = df.iloc[test_idx]
    train_val_df = df.iloc[train_val_idx]

    train_df, val_df = train_test_split(
        train_val_df, test_size=0.1,
        stratify=train_val_df['label_idx'], random_state=SEED
    )

    train_ds   = build_dataset(train_df, is_training=True)
    val_ds     = build_dataset(val_df,   is_training=False)
    test_ds    = build_dataset(test_df,  is_training=False)
    predict_ds = build_predict_dataset(test_df)

    model = build_attention_cnn_model()

    checkpoint_path = f'/kaggle/working/best_cbam_cnn_fold_{fold_num}.keras'
    checkpoint      = ModelCheckpoint(
        checkpoint_path, monitor='val_accuracy',
        save_best_only=True, mode='max', verbose=1
    )
    early_stopping  = EarlyStopping(
        monitor='val_accuracy', patience=10,
        restore_best_weights=True, verbose=1
    )

    print("\nвЏі Training Model...")
    history = model.fit(
        train_ds, validation_data=val_ds,
        epochs=EPOCHS, callbacks=[checkpoint, early_stopping], verbose=1
    )

    # --- Plot Training Curves ---
    print(f"\nрџ“Љ Generating Training Curves for Fold {fold_num}...")
    plt.figure(figsize=(14, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'],     label='Training Accuracy',   linewidth=2)
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
    plt.title(f'Fold {fold_num}: CBAM CNN Accuracy')
    plt.xlabel('Epoch'); plt.ylabel('Accuracy')
    plt.legend(loc='lower right')
    plt.grid(True, linestyle='--', alpha=0.7)

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'],     label='Training Loss',   linewidth=2)
    plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
    plt.title(f'Fold {fold_num}: CBAM CNN Loss')
    plt.xlabel('Epoch'); plt.ylabel('Loss')
    plt.legend(loc='upper right')
    plt.grid(True, linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.savefig(f'/kaggle/working/cbam_cnn_training_curves_fold_{fold_num}.png', dpi=300)
    plt.close()

    # --- Evaluate on Test Set ---
    print("\nрџ”Ќ Evaluating on unseen Pure Test Set...")
    model.load_weights(checkpoint_path)

    test_loss, test_acc = model.evaluate(test_ds, verbose=0)
    fold_accuracies.append(test_acc)

    # --- Advanced Metrics ---
    true_labels = test_df['label_idx'].values
    pred_probs  = model.predict(predict_ds, verbose=0)
    pred_labels = np.argmax(pred_probs, axis=1)

    f1   = f1_score(true_labels, pred_labels, average='weighted')
    prec = precision_score(true_labels, pred_labels, average='weighted')
    rec  = recall_score(true_labels, pred_labels, average='weighted')
    cm   = confusion_matrix(true_labels, pred_labels)

    fold_f1_scores.append(f1)
    fold_precisions.append(prec)
    fold_recalls.append(rec)
    fold_conf_matrices.append(cm)

    # --- Plot Confusion Matrix ---
    print(f"\nрџ“Љ Generating Confusion Matrix for Fold {fold_num}...")
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Fold {fold_num}: CBAM CNN Confusion Matrix')
    plt.xlabel('Predicted Label'); plt.ylabel('True Label')
    plt.tight_layout()
    plt.savefig(f'/kaggle/working/cbam_cnn_confusion_matrix_fold_{fold_num}.png', dpi=300)
    plt.close()

    print(f"вњ… Fold {fold_num} | Acc: {test_acc*100:.2f}% | F1: {f1:.4f} | Precision: {prec:.4f} | Recall: {rec:.4f}")

    # --- Memory Cleanup ---
    print("рџ§№ Wiping GPU Memory clean for the next fold...")
    del model, train_ds, val_ds, test_ds, predict_ds, history
    tf.keras.backend.clear_session()
    gc.collect()

# ==========================================
# 8. SAVE FINAL ARRAYS
# ==========================================
np.save('/kaggle/working/AttentionCNN_5Fold_Accuracies_DualGPU.npy',  np.array(fold_accuracies))
np.save('/kaggle/working/AttentionCNN_5Fold_F1_DualGPU.npy',          np.array(fold_f1_scores))
np.save('/kaggle/working/AttentionCNN_5Fold_Precision_DualGPU.npy',   np.array(fold_precisions))
np.save('/kaggle/working/AttentionCNN_5Fold_Recall_DualGPU.npy',      np.array(fold_recalls))
np.save('/kaggle/working/AttentionCNN_5Fold_ConfMatrix_DualGPU.npy',  np.array(fold_conf_matrices))

print(f"\nрџЋ‰ ALL {K_FOLDS} FOLDS COMPLETE!")
print(f"рџЏ† Avg Accuracy  : {np.mean(fold_accuracies)*100:.2f}% (+/- {np.std(fold_accuracies)*100:.2f}%)")
print(f"рџЏ† Avg F1 Score  : {np.mean(fold_f1_scores):.4f} (+/- {np.std(fold_f1_scores):.4f})")
print(f"рџЏ† Avg Precision : {np.mean(fold_precisions):.4f} (+/- {np.std(fold_precisions):.4f})")
print(f"рџЏ† Avg Recall    : {np.mean(fold_recalls):.4f} (+/- {np.std(fold_recalls):.4f})")
print(f"вЏ±пёЏ  Total Experiment Time: {(time.time() - start_time) / 60:.2f} minutes")
```
