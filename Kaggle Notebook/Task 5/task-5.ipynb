```python
import os
import random
import time
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
import cv2

from tensorflow.keras.layers import (
    Input, Conv2D, MaxPooling2D, BatchNormalization, Activation,
    Dense, Dropout, GlobalAveragePooling2D, GlobalMaxPooling2D,
    Reshape, Multiply, Add, Concatenate, Lambda
)
from tensorflow.keras.models import Model
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, precision_score, recall_score, confusion_matrix

# =========================================================================
# ⚠️ USER CONFIGURATION
# =========================================================================
SAVED_MODEL_PATH = '/kaggle/input/datasets/humairahabibusha/cbam-keras-file/best_cbam_cnn_fold_5.keras' 
DATA_DIR_1 = '/kaggle/input/datasets/uraninjo/augmented-alzheimer-mri-dataset-v2'
DATA_DIR_2 = '/kaggle/input/datasets/ninadaithal/imagesoasis' 
# =========================================================================

# Scientific Locks
SEED = 42
os.environ['PYTHONHASHSEED'] = str(SEED)
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

try:
    strategy = tf.distribute.MirroredStrategy()
except Exception:
    strategy = tf.distribute.get_strategy()

BATCH_SIZE = 32 * strategy.num_replicas_in_sync
AUTOTUNE = tf.data.AUTOTUNE
class_names = ['MildDemented', 'ModerateDemented', 'NonDemented', 'VeryMildDemented']
class_indices = {name: idx for idx, name in enumerate(class_names)}


# =========================================================================
# ARCHITECTURE REDEFINITION
# =========================================================================
def channel_attention(x, ratio=8):
    channels = x.shape[-1]
    shared_dense_1 = Dense(channels // ratio, activation='relu',  use_bias=False)
    shared_dense_2 = Dense(channels,          activation='linear', use_bias=False)
    
    avg_pool = GlobalAveragePooling2D()(x)
    avg_pool = Reshape((1, 1, channels))(avg_pool)
    avg_out  = shared_dense_1(avg_pool)
    avg_out  = shared_dense_2(avg_out)
    
    max_pool = GlobalMaxPooling2D()(x)
    max_pool = Reshape((1, 1, channels))(max_pool)
    max_out  = shared_dense_1(max_pool)
    max_out  = shared_dense_2(max_out)
    
    attention = Add()([avg_out, max_out])
    attention = Activation('sigmoid')(attention)
    return Multiply()([x, attention])

def spatial_attention(x, kernel_size=7):
    avg_pool = Lambda(lambda t: tf.reduce_mean(t, axis=-1, keepdims=True), 
                      output_shape=lambda s: s[:-1] + (1,))(x)
    max_pool = Lambda(lambda t: tf.reduce_max(t,  axis=-1, keepdims=True), 
                      output_shape=lambda s: s[:-1] + (1,))(x)
    concat    = Concatenate(axis=-1)([avg_pool, max_pool])
    attention = Conv2D(1, kernel_size, padding='same', use_bias=False)(concat)
    attention = Activation('sigmoid')(attention)
    return Multiply()([x, attention])

def cbam_block(x, ratio=8, kernel_size=7):
    x = channel_attention(x, ratio=ratio)
    x = spatial_attention(x, kernel_size=kernel_size)
    return x

# 🚀 FIX: Added 'use_strategy' toggle to prevent batch_size=0 crashes during inference
def build_attention_cnn_model(use_strategy=True):
    tf.keras.backend.clear_session()
    
    def _build_model():
        inputs = Input(shape=(224, 224, 3))
        
        x = Conv2D(32, (3, 3), padding='same')(inputs)
        x = BatchNormalization()(x); x = Activation('relu')(x)
        x = MaxPooling2D(pool_size=(2, 2))(x); x = cbam_block(x)
        
        x = Conv2D(64, (3, 3), padding='same')(x)
        x = BatchNormalization()(x); x = Activation('relu')(x)
        x = MaxPooling2D(pool_size=(2, 2))(x); x = cbam_block(x)
        
        x = Conv2D(128, (3, 3), padding='same')(x)
        x = BatchNormalization()(x); x = Activation('relu')(x)
        x = MaxPooling2D(pool_size=(2, 2))(x); x = cbam_block(x)
        
        x = Conv2D(256, (3, 3), padding='same')(x)
        x = BatchNormalization()(x); x = Activation('relu')(x)
        x = MaxPooling2D(pool_size=(2, 2))(x); x = cbam_block(x)
        
        x = GlobalAveragePooling2D()(x)
        x = Dense(256, activation='relu')(x)
        x = Dropout(0.5)(x)
        outputs = Dense(len(class_names), activation='softmax')(x)
        
        model = Model(inputs=inputs, outputs=outputs)
        model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                      loss='categorical_crossentropy', metrics=['accuracy'])
        return model

    if use_strategy:
        with strategy.scope():
            return _build_model()
    else:
        return _build_model()


# =========================================================================
# TASK 5.1: EXPLAINABILITY USING GRAD-CAM
# =========================================================================
print("\n" + "="*50)
print("🔍 TASK 5.1: EXPLAINABILITY (Grad-CAM)")
print("="*50)

def get_img_array(img_path, size=(224, 224)):
    img = tf.keras.preprocessing.image.load_img(img_path, target_size=size)
    array = tf.keras.preprocessing.image.img_to_array(img)
    array = np.expand_dims(array, axis=0)
    return array / 255.0

def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    grad_model = tf.keras.models.Model(model.inputs, [model.get_layer(last_conv_layer_name).output, model.output])
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]
    grads = tape.gradient(class_channel, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

def display_gradcam(img_path, heatmap, pred_label, true_label, alpha=0.4):
    # 1. Read and convert to RGB immediately (fixes weird color overlaps)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (224, 224))
    
    # 2. Rescale heatmap to 0-255
    heatmap = np.uint8(255 * heatmap)
    
    # 3. Create the RGB heatmap using the 'jet' colormap
    jet = cm.get_cmap("jet")
    jet_colors = jet(np.arange(256))[:, :3] # Values from 0.0 to 1.0
    jet_heatmap = jet_colors[heatmap]
    jet_heatmap = cv2.resize(jet_heatmap, (img.shape[1], img.shape[0]))
    jet_heatmap = jet_heatmap * 255.0 # Scale to 0-255 to match the original image
    
    # 4. Math operations on NumPy arrays (fixes the TypeError!)
    superimposed_img = jet_heatmap * alpha + img
    superimposed_img = np.clip(superimposed_img, 0, 255).astype(np.uint8)
    
    # 5. Display
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 3, 1)
    plt.imshow(img); plt.title(f"Original\nTrue: {true_label}"); plt.axis('off')
    
    plt.subplot(1, 3, 2)
    plt.imshow(heatmap, cmap='jet'); plt.title("Grad-CAM Heatmap"); plt.axis('off')
    
    plt.subplot(1, 3, 3)
    plt.imshow(superimposed_img); plt.title(f"Superimposed\nPred: {pred_label}"); plt.axis('off')
    
    color = 'green' if pred_label == true_label else 'red'
    ax = plt.gca()
    for spine in ax.spines.values():
        spine.set_visible(True)
        spine.set_edgecolor(color)
        spine.set_linewidth(5)
    plt.tight_layout()
    plt.show()

print("Building fresh model and loading weights (Single-GPU mode)...")
# 🚀 FIX: We set use_strategy=False here so feeding 1 image doesn't crash the dual-GPUs
best_model = build_attention_cnn_model(use_strategy=False)
best_model.load_weights(SAVED_MODEL_PATH)

last_conv_layer = None
for layer in reversed(best_model.layers):
    if isinstance(layer, tf.keras.layers.Conv2D):
        last_conv_layer = layer.name
        break
print(f"✅ Target Conv Layer for Grad-CAM: {last_conv_layer}")

sample_paths, sample_labels = [], []
for root, dirs, files in os.walk(DATA_DIR_1):
    for f in files[:30]: 
        if f.endswith(('.jpg', '.png', '.jpeg')):
            sample_paths.append(os.path.join(root, f))
            sample_labels.append(os.path.basename(root))

found_correct, found_incorrect = False, False
print("\nGenerating Grad-CAM Visualizations...")
for path, true_label in zip(sample_paths, sample_labels):
    if found_correct and found_incorrect: break
    img_array = get_img_array(path)
    preds = best_model.predict(img_array, verbose=0)
    pred_label = class_names[np.argmax(preds[0])]
    is_correct = (pred_label == true_label)
    
    if is_correct and not found_correct:
        print("-> Found Correct Prediction Sample:")
        heatmap = make_gradcam_heatmap(img_array, best_model, last_conv_layer)
        display_gradcam(path, heatmap, pred_label, true_label)
        found_correct = True
    elif not is_correct and not found_incorrect:
        print("-> Found Incorrect Prediction Sample:")
        heatmap = make_gradcam_heatmap(img_array, best_model, last_conv_layer)
        display_gradcam(path, heatmap, pred_label, true_label)
        found_incorrect = True


# =========================================================================
# TASK 5.2: GENERALIZABILITY TESTING 
# =========================================================================
print("\n" + "="*50)
print("🌍 TASK 5.2: GENERALIZABILITY TESTING")
print("="*50)

filepaths_2, labels_2 = [], []
for root, dirs, files in os.walk(DATA_DIR_2):
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            base_label = os.path.basename(root).lower().replace("_", "")
            mapped_label = None
            if "verymild" in base_label: mapped_label = "VeryMildDemented"
            elif "non" in base_label: mapped_label = "NonDemented"
            elif "moderate" in base_label: mapped_label = "ModerateDemented"
            elif "mild" in base_label: mapped_label = "MildDemented"
            
            if mapped_label:
                filepaths_2.append(os.path.join(root, file))
                labels_2.append(mapped_label)

df_2 = pd.DataFrame({'filepath': filepaths_2, 'label': labels_2})
if df_2.empty:
    print(f"⚠️ Could not load second dataset. Please verify DATA_DIR_2 path: {DATA_DIR_2}")
else:
    df_2['label_idx'] = df_2['label'].map(class_indices)
    print(f"✅ Second Dataset Loaded: {len(df_2)} valid images found.")

    train_val_df_2, test_df_2 = train_test_split(df_2, test_size=0.1, stratify=df_2['label_idx'], random_state=SEED)
    train_df_2, val_df_2 = train_test_split(train_val_df_2, test_size=0.1111, stratify=train_val_df_2['label_idx'], random_state=SEED)

    def decode_image(filepath, label):
        image_bytes = tf.io.read_file(filepath)
        image = tf.io.decode_image(image_bytes, channels=3, expand_animations=False)
        image = tf.image.resize(image, [224, 224])
        image = image / 255.0
        label = tf.one_hot(label, depth=len(class_names))
        return image, label

    def build_dataset(dataframe, is_training=False):
        dataset = tf.data.Dataset.from_tensor_slices((dataframe['filepath'].values, dataframe['label_idx'].values))
        
        # Shuffle BEFORE mapping saves memory
        if is_training: 
            dataset = dataset.shuffle(buffer_size=2048, seed=SEED)
            
        dataset = dataset.map(decode_image, num_parallel_calls=AUTOTUNE)
        
        # 🚀 FIX: Removed dataset.cache() so Kaggle doesn't run out of RAM!
        return dataset.batch(BATCH_SIZE).prefetch(AUTOTUNE)

    train_ds_2 = build_dataset(train_df_2, is_training=True)
    val_ds_2   = build_dataset(val_df_2, is_training=False)
    test_ds_2  = build_dataset(test_df_2, is_training=False)

    print("\n⏳ Retraining Model on Second Dataset...")
    # 🚀 FIX: We set use_strategy=True here to utilize Dual-GPUs for full dataset retraining
    gen_model = build_attention_cnn_model(use_strategy=True)
    
    checkpoint_gen = tf.keras.callbacks.ModelCheckpoint(
        '/kaggle/working/generalizability_cbam_cnn.keras', monitor='val_accuracy',
        save_best_only=True, mode='max', verbose=1
    )
    early_stopping_gen = tf.keras.callbacks.EarlyStopping(
        monitor='val_accuracy', patience=10, restore_best_weights=True, verbose=1
    )

    gen_start_time = time.time()
    history_gen = gen_model.fit(train_ds_2, validation_data=val_ds_2, epochs=50, callbacks=[checkpoint_gen, early_stopping_gen], verbose=1)

    # --- Plot Generalizability Curves ---
    print("\n📊 Generating Training Curves for Dataset 2...")
    plt.figure(figsize=(14, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history_gen.history['accuracy'], label='Training Accuracy', linewidth=2)
    plt.plot(history_gen.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
    plt.title('Generalizability Retraining: Accuracy'); plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend(); plt.grid(True, linestyle='--', alpha=0.7)
    
    plt.subplot(1, 2, 2)
    plt.plot(history_gen.history['loss'], label='Training Loss', linewidth=2)
    plt.plot(history_gen.history['val_loss'], label='Validation Loss', linewidth=2)
    plt.title('Generalizability Retraining: Loss'); plt.xlabel('Epoch'); plt.ylabel('Categorical Crossentropy'); plt.legend(); plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.savefig('/kaggle/working/generalizability_training_curves.png', dpi=300)
    plt.show()

    # --- Evaluate ---
    print("\n🔍 Evaluating Generalizability on Second Dataset Test Split...")
    gen_model.load_weights('/kaggle/working/generalizability_cbam_cnn.keras')
    test_loss_2, test_acc_2 = gen_model.evaluate(test_ds_2, verbose=0)
    
    true_labels_2 = test_df_2['label_idx'].values
    pred_probs_2  = gen_model.predict(test_ds_2, verbose=0)
    pred_labels_2 = np.argmax(pred_probs_2, axis=1)

    f1_2   = f1_score(true_labels_2, pred_labels_2, average='weighted')
    prec_2 = precision_score(true_labels_2, pred_labels_2, average='weighted')
    rec_2  = recall_score(true_labels_2, pred_labels_2, average='weighted')
    cm_2   = confusion_matrix(true_labels_2, pred_labels_2)

    print("\n" + "="*40)
    print("📊 GENERALIZABILITY ANALYSIS RESULTS")
    print("="*40)
    print(f"Dataset 2 Accuracy      : {test_acc_2*100:.2f}%")
    print(f"Dataset 2 F1 Score      : {f1_2:.4f}")
    print(f"Dataset 2 Precision     : {prec_2:.4f}")
    print(f"Dataset 2 Recall        : {rec_2:.4f}")
    print(f"Dataset 2 Training Time : {(time.time() - gen_start_time)/60:.2f} mins")
    
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_2, annot=True, fmt='d', cmap='Oranges', xticklabels=class_names, yticklabels=class_names)
    plt.title('Generalizability Test: Confusion Matrix on Dataset 2')
    plt.xlabel('Predicted Label'); plt.ylabel('True Label'); plt.tight_layout()
    plt.savefig('/kaggle/working/generalizability_confusion_matrix.png', dpi=300)
    plt.show()
```
