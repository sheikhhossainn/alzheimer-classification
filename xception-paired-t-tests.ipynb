```python
import numpy as np
from scipy.stats import ttest_rel
import os

# ============================================================
# FINAL RESEARCH STEP: 3-WAY PAIRED T-TEST MATRIX
# ============================================================

# 1. Exact file paths using your provided Kaggle dataset path
vgg16_path = '/kaggle/input/datasets/shahiismyname/xception-paired-t-test/vgg16_correctness.npy'
vgg19_path = '/kaggle/input/datasets/shahiismyname/xception-paired-t-test/vgg19_correctness.npy'
xception_path = '/kaggle/input/datasets/shahiismyname/xception-paired-t-test/xception_correctness.npy'

print("Locating files...")
if os.path.exists(vgg16_path) and os.path.exists(vgg19_path) and os.path.exists(xception_path):
    print("All three correctness files found.\n")
    
    # Load the 1s and 0s arrays
    vgg16_res = np.load(vgg16_path)
    vgg19_res = np.load(vgg19_path)
    xcep_res = np.load(xception_path)

    # Calculate final pure test accuracies
    acc_16 = np.mean(vgg16_res) * 100
    acc_19 = np.mean(vgg19_res) * 100
    acc_xc = np.mean(xcep_res) * 100

    print("-" * 65)
    print("FINAL MODEL ACCURACIES (Unseen Test Set)")
    print("-" * 65)
    print(f"VGG16 Final Accuracy:    {acc_16:.2f}%")
    print(f"VGG19 Final Accuracy:    {acc_19:.2f}%")
    print(f"Xception Final Accuracy: {acc_xc:.2f}%")
    print("-" * 65 + "\n")

    # Define the pairings for the matrix
    comparisons = [
        ("VGG16", "VGG19", vgg16_res, vgg19_res, acc_16, acc_19),
        ("VGG16", "Xception", vgg16_res, xcep_res, acc_16, acc_xc),
        ("VGG19", "Xception", vgg19_res, xcep_res, acc_19, acc_xc)
    ]

    print("-" * 65)
    print("PAIRWISE T-TEST RESULTS (Full Matrix)")
    print("-" * 65)
    # Table Header
    print(f"{'Comparison':<20} | {'T-Statistic':<12} | {'P-Value':<10} | {'Superior Model'}")
    print("-" * 65)

    # Run the tests and print a clean summary table
    for name1, name2, res1, res2, a1, a2 in comparisons:
        t_stat, p_val = ttest_rel(res1, res2)
        
        comp_name = f"{name1} vs {name2}"
        
        # Determine winner
        if p_val < 0.05:
            winner = name1 if a1 > a2 else name2
        else:
            winner = "None (Equal)"
            
        # Format p-value nicely for research (e.g., < 0.001 instead of 0.000000)
        p_val_str = "< 0.001" if p_val < 0.001 else f"{p_val:.4f}"
        
        print(f"{comp_name:<20} | {t_stat:<12.4f} | {p_val_str:<10} | {winner}")

    print("-" * 65)

else:
    print("ERROR: One or more .npy files are still missing! Please double-check the folder structure.")
```
